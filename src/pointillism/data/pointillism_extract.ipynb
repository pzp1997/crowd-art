```python
import json
```

```python
data_lines = []

with open('pointillism.log', 'r') as f:
    for line in f:
        line = line.strip()
        if 'DATA: ' in line and 'OVERRIDE' not in line:
            data_lines.append(line[line.index('DATA: ') + len('DATA: '):])

data_lines          
```

```python
len(data_lines)
```

```python
data_lines[0]
```

```python
# c, r, color, timestamp
data = [
    set(tuple(point.values()) for point in json.loads(line))
    for line in data_lines
]
```

```python
data_maximal = [
    sorted(row, key=lambda x: x[3])
    for row in data
    if not any(row < other for other in data)
]
```

```python
data_maximal
```

```python
with open('pointillism-peers.json', 'w') as f:
    json.dump(data_maximal, f)
```

```python
with open('pointillism-peers.json', 'r') as f:
    data_maximal = json.load(f)
```
